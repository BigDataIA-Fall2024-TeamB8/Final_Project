```python
from diagrams import Cluster, Diagram
from diagrams.onprem.workflow import Airflow
from diagrams.onprem.database import PostgreSQL
from diagrams.onprem.client import User
from diagrams.onprem.container import Docker
from diagrams.onprem.inmemory import Redis
from diagrams.aws.storage import S3
from diagrams.custom import Custom
from diagrams.generic.compute import Rack as SwaggerUI
from diagrams.generic.compute import Rack as JWTAuth
from diagrams.programming.framework import Fastapi


ted_icon = "/content/ted_icon.png"
youtube_icon = "/content/youtube.png"
pinecone_icon = "/content/Pinecone.png"
langgraph_icon = "/content/Langraph.png"
snowflake_icon = "/content/Snowflake.png"
openai_icon = "/content/chat-gpt.png"
streamlit_icon_path = "/content/streamlit.png"
selenium_icon_path = "/content/Selenium.png"

with Diagram("TED Talk Query Platform Architecture", show=True, direction="TB"):

    user = User("User")

    with Cluster("Docker Container 1"):
        ted = Custom("TED API", ted_icon)
        youtube = Custom("YouTube API", youtube_icon)
        airflow = Airflow("Airflow")
        s3_bucket = S3("S3 Bucket")
        selenium = Custom("Selenium", selenium_icon_path)

        ted >> airflow
        youtube >> airflow
        selenium >> airflow
        airflow >> s3_bucket

    with Cluster("Docker Container 2"):
        sql_db = PostgreSQL("SQLCredential Database")
        pinecone = Custom("Pinecone Vector Database", pinecone_icon)
        snowflake = Custom("Snowflake", snowflake_icon)
        fastapi = Fastapi("FastAPI")
        swagger_ui = SwaggerUI("Swagger UI")
        jwt_auth = JWTAuth("JWT Auth")
        langgraph = Custom("LangGraph", langgraph_icon)
        ui = Custom("Streamlit UI", streamlit_icon_path)
        openai = Custom("OpenAI", openai_icon)

        s3_bucket >> fastapi
        s3_bucket >> snowflake
        s3_bucket >> pinecone
        pinecone >> fastapi
        snowflake >> fastapi
        fastapi >> sql_db
        fastapi >> pinecone
        fastapi >> jwt_auth
        fastapi >> swagger_ui
        fastapi >> langgraph
        langgraph >> ui
        langgraph >> openai
        openai >> ui
        ui >> user
```
